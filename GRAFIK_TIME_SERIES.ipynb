# **PERIODE BULANAN**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data dari Excel (file_path sudah diberikan sebelumnya)
file_path = "/content/drive/MyDrive/TUGAS AKHIR/DATA JADI/DATA HASIL RATA RATA /GRAFIK TIME SERIES/periode bulanan.xlsx"

# Membaca file Excel
data = pd.read_excel(file_path)

# Asumsi: Data memiliki 12 entri, masing-masing mewakili satu bulan
months = [
    "Januari", "Februari", "Maret", "April", "Mei", "Juni",
    "Juli", "Agustus", "September", "Oktober", "November", "Desember"
]
data.index = months  # Menambahkan nama bulan sebagai indeks

# Plot time series untuk setiap stasiun
plt.figure(figsize=(12, 6))
for column in data.columns:
    plt.plot(data.index, data[column], marker='o', label=column)

# Menambahkan grid
plt.grid(visible=True, which='both', linestyle='--', linewidth=0.5, alpha=0.7)

# Menambahkan judul, label, dan legend
plt.title("", fontsize=14)
plt.xlabel("", fontsize=12)
plt.ylabel("Periode (s)", fontsize=12)
plt.legend(title="Lokasi", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()

# Menampilkan plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca file Excel pertama: periode tahunan (SWH)
file_path_swh = "/content/drive/MyDrive/TUGAS AKHIR/DATA JADI/DATA HASIL RATA RATA /GRAFIK TIME SERIES/periode tahunan.xlsx"
data_swh = pd.read_excel(file_path_swh)

# Membaca file Excel kedua: wave power tahunan
file_path_wave_power = "/content/drive/MyDrive/TUGAS AKHIR/DATA JADI/DATA HASIL RATA RATA /GRAFIK TIME SERIES/wave power tahunan.xlsx"
data_wave_power = pd.read_excel(file_path_wave_power)

# Asumsi: Data mewakili periode tahunan (2019-2023)
years = ["2019", "2020", "2021", "2022", "2023"]
data_swh.index = years  # Menambahkan tahun sebagai indeks untuk SWH
data_wave_power.index = years  # Menambahkan tahun sebagai indeks untuk Wave Power

# Plot grafik untuk SWH
plt.figure(figsize=(12, 6))
for column in data_swh.columns:
    plt.plot(data_swh.index, data_swh[column], marker='o', label=column)

# Menambahkan grid, judul, dan label untuk SWH
plt.grid(visible=True, which='both', linestyle='--', linewidth=0.5, alpha=0.7)
plt.title("", fontsize=14)
plt.xlabel("", fontsize=12)
plt.ylabel("Periode (s)", fontsize=12)
plt.legend(title="Lokasi", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot grafik untuk Wave Power
plt.figure(figsize=(12, 6))
for column in data_wave_power.columns:
    plt.plot(data_wave_power.index, data_wave_power[column], marker='o', label=column)

# Menambahkan grid, judul, dan label untuk Wave Power
plt.grid(visible=True, which='both', linestyle='--', linewidth=0.5, alpha=0.7)
plt.title("", fontsize=14)
plt.xlabel("", fontsize=12)
plt.ylabel("Wave Power (kW/m)", fontsize=12)
plt.legend(title="Lokasi", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
